```python
from docx import Document
from icecream import ic
import pandas as pd
```

```python
document = Document("Документ_предоставлен_КонсультантПлюс.docx")
```

```python
df = pd.DataFrame()

```

```python
for table in document.tables:

    # Проход по всем строкам в текущей таблице
    for row in table.rows:
        # Создаем список с текстом из ячеек строки
        row_data = [cell.text.strip() for cell in row.cells]
        
        # Преобразуем строку в DataFrame
        row_df = pd.DataFrame([row_data])

        # Объединяем текущую строку с общим DataFrame
        df = pd.concat([df, row_df], ignore_index=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.columns=['num', 'prior', 'adm', 'city', 'other']
```

```python
to_rem = ['июн', 'ред', 'вед', 'ключ','п/п']
```

```python
for i in to_rem:
    df =df[~df['prior'].str.contains(i)]
    
df.query('prior != ""', inplace=True)
```
